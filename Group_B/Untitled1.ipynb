```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.isna().sum()
# No null values, it's clean
```

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='age', data=df)
plt.title('Count Plot of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='age', data=df)
plt.title('Violin Plot of Age by Sex')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='thalachh', y='chol', data=df)
plt.title('Scatter Plot of Thalachh vs Cholesterol')
plt.xlabel('Thalachh')
plt.ylabel('Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='cp', y='thalachh', data=df)
plt.title('Swarm Plot of Chest Pain Type vs Thalachh')
plt.xlabel('Chest Pain Type')
plt.ylabel('Thalachh')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='output', y='thalachh', data=df)
plt.title('Boxplot of Output vs Thalachh')
plt.xlabel('Output')
plt.ylabel('Thalachh')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='cp', y='output', data=df)
plt.title('Barplot of Chest Pain Type vs Output')
plt.xlabel('Chest Pain Type')
plt.ylabel('Output')
plt.show()
```

```python
correlation_matrix = df.corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df_sorted = df.sort_values(by='age')
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='thalachh', data=df_sorted)
plt.title('Line Plot of Age vs Max Heart Rate')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='chol', y='sex', data=df_sorted)
plt.title('Line Plot of chol vs sex')
plt.xlabel('chol')
plt.ylabel('sex')
plt.show()
```

```python
sns.histplot(data=df,x=df.trtbps, hue='output')
```

```python
temp_df = df[['age','cp', 'trtbps','chol','fbs','oldpeak','output']]
plt.figure(figsize=(15,10))
sns.pairplot(temp_df,hue="output")
plt.title("Looking for Insites in Data")
plt.legend("HeartDisease")
plt.tight_layout()
plt.plot()
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(temp_df.columns,1):
    plt.subplot(4,3,i)
    plt.title(f"Distribution of {col} Data")
    sns.histplot(df[col],kde=True)
    plt.tight_layout()
    plt.plot()
```
